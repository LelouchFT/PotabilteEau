```python
# Importations
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import cross_val_score, learning_curve, validation_curve
from sklearn.metrics import classification_report, confusion_matrix, recall_score, accuracy_score, make_scorer
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

# Custom Transformer pour ajouter du bruit
class NoiseAdder(BaseEstimator, TransformerMixin):
    def __init__(self, noise_level=0.01):
        self.noise_level = noise_level

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X_noisy = X.copy()
        numeric_cols = pd.DataFrame(X_noisy).select_dtypes(include=[np.number]).columns
        noise = np.random.normal(0.0, self.noise_level, size=X_noisy[numeric_cols].shape)
        X_noisy[numeric_cols] += noise
        return X_noisy

# ModÃ¨les de base
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42
)

xgb = XGBClassifier(
    learning_rate=0.1,
    max_depth=6,
    n_estimators=100,
    subsample=0.8,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# VotingClassifier
voting = VotingClassifier(
    estimators=[('rf', rf), ('xgb', xgb)],
    voting='soft',
    weights=[1, 2],
    n_jobs=-1
)

# Pipeline complet
pipeline = ImbPipeline(steps=[
    ('feature_selection', SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42), threshold="mean")),
    ('smote', SMOTE(random_state=42)),
    ('noise', NoiseAdder(noise_level=0.01)),
    ('classifier', voting)
])

# 1. EntraÃ®nement
pipeline.fit(X_train, y_train)

# 2. PrÃ©diction
y_pred = pipeline.predict(X_test)

# 3. Sauvegarde du modÃ¨le
joblib.dump(pipeline, 'pipeline_voting.joblib')

# 4. Ã‰valuation
print("\n--- Ã‰valuation du modÃ¨le ---")
print("Score recall :", recall_score(y_test, y_pred, average='macro'))
print("Score accuracy :", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 5. Matrice de confusion
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("PrÃ©dit")
plt.ylabel("RÃ©el")
plt.title("Matrice de confusion")
plt.show()

# 6. Courbe dâ€™apprentissage
recall_scorer = make_scorer(recall_score, average='macro')
train_sizes, train_scores, test_scores = learning_curve(
    pipeline, X_train, y_train,
    cv=5,
    scoring=recall_scorer,
    n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10)
)

train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(8, 5))
plt.plot(train_sizes, train_scores_mean, 'o-', label="Train")
plt.plot(train_sizes, test_scores_mean, 'o-', label="Test")
plt.xlabel("Taille de l'ensemble d'entraÃ®nement")
plt.ylabel("Recall (macro)")
plt.title("Courbe d'apprentissage - Pipeline VotingClassifier")
plt.legend()
plt.grid(True)
plt.show()

# 7. Validation Curve sur RandomForest (n_estimators)
param_range = [50, 100, 150, 200, 250, 300]

train_scores_val, test_scores_val = validation_curve(
    RandomForestClassifier(random_state=42),
    X_train, y_train,
    param_name="n_estimators",
    param_range=param_range,
    cv=5,
    scoring="recall_macro",
    n_jobs=-1
)

train_scores_mean_val = np.mean(train_scores_val, axis=1)
test_scores_mean_val = np.mean(test_scores_val, axis=1)

plt.figure(figsize=(8, 5))
plt.plot(param_range, train_scores_mean_val, 'o-', label="Train")
plt.plot(param_range, test_scores_mean_val, 'o-', label="Validation")
plt.xlabel("n_estimators (RandomForest)")
plt.ylabel("Recall (macro)")
plt.title("Validation Curve - RandomForest")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Importations
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import classification_report, confusion_matrix, recall_score, accuracy_score, make_scorer
from sklearn.model_selection import learning_curve
from imblearn.over_sampling import SMOTE

# Fonction d'ajout de bruit (augmentation)
def add_noise(X, noise_level=0.03):
    X_noisy = X.copy()
    numeric_cols = X.select_dtypes(include=[np.number]).columns
    noise = np.random.normal(loc=0.0, scale=noise_level, size=X[numeric_cols].shape)
    X_noisy[numeric_cols] += noise
    return X_noisy

# 1. SÃ©lection des meilleures features
rf_selector = RandomForestClassifier(n_estimators=100, random_state=42)
rf_selector.fit(X_train, y_train)

selector = SelectFromModel(rf_selector, prefit=True, threshold="mean")
X_train_sel = selector.transform(X_train)
X_test_sel = selector.transform(X_test)

selected_columns = X_train.columns[selector.get_support()]
print(f"Nombre de features sÃ©lectionnÃ©es : {len(selected_columns)}")
print("Features sÃ©lectionnÃ©es :", selected_columns.tolist())

# 2. RÃ©Ã©quilibrage des classes avec SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_sel, y_train)

# Convertir X_train_resampled en DataFrame pandas
X_train_resampled = pd.DataFrame(X_train_resampled, columns=selected_columns)

# VÃ©rification taille aprÃ¨s SMOTE
print(f"Taille initiale de X_train_sel : {X_train_sel.shape[0]}")
print(f"Taille aprÃ¨s SMOTE : {X_train_resampled.shape[0]}")

# 3. GÃ©nÃ©ration d'augmentations pour atteindre environ 4x (1 original + 3 bruitÃ©s)
X_augmented_list = [X_train_resampled]  # Commence avec original
y_augmented_list = [pd.Series(y_train_resampled)]

for _ in range(3):  # 3 bruitages diffÃ©rents
    X_noisy = add_noise(X_train_resampled, noise_level=0.03)
    X_augmented_list.append(X_noisy)
    y_augmented_list.append(pd.Series(y_train_resampled))

# ConcatÃ©nation
X_train_augmented = pd.concat(X_augmented_list, ignore_index=True)
y_train_augmented = pd.concat(y_augmented_list, ignore_index=True)

print(f"Taille finale aprÃ¨s augmentation : {X_train_augmented.shape[0]}")

# 4. ModÃ¨le avec RandomForest
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=6,
    min_samples_leaf=10,
    random_state=42
)

# 5. EntraÃ®nement
rf.fit(X_train_augmented, y_train_augmented)
y_pred = rf.predict(X_test_sel)

# Sauvegarde du modÃ¨le
joblib.dump(rf, 'modele_rf_augmented_x3.joblib')

# 6. Ã‰valuation
print("Recall macro :", recall_score(y_test, y_pred, average='macro'))
print("Accuracy :", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 7. Matrice de confusion
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("PrÃ©dit")
plt.ylabel("RÃ©el")
plt.title("Matrice de confusion")
plt.show()

# 8. Courbe d'apprentissage
recall_scorer = make_scorer(recall_score, average='macro')
train_sizes, train_scores, test_scores = learning_curve(
    rf, X_train_augmented, y_train_augmented,
    cv=5,
    scoring=recall_scorer,
    n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10)
)

train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(8, 5))
plt.plot(train_sizes, train_scores_mean, 'o-', label="Train")
plt.plot(train_sizes, test_scores_mean, 'o-', label="Test")
plt.xlabel("Taille de l'ensemble d'entraÃ®nement")
plt.ylabel("Recall (macro)")
plt.title("Courbe d'apprentissage aprÃ¨s augmentation x3 avec RandomForest")
plt.legend()
plt.grid(True)
plt.show()
```
